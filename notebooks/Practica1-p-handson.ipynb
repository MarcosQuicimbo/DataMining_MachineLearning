2) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column.

```python
import random

# Set the dimensions of the array
rows = 3
columns = 4

# Generate random integers and store them in the array
random_array = [[random.randint(1, 100) for _ in range(columns)] for _ in range(rows)]

# Print the array
for row in random_array:
    print(row)

# Calculate the sum of each row
row_sums = [sum(row) for row in random_array]

# Calculate the sum of each column
column_sums = [sum(column) for column in zip(*random_array)]

# Print the sums of each row
print("Sum of each row:", row_sums)

# Print the sums of each column
print("Sum of each column:", column_sums)
```

4) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame

```python
import pandas as pd

# Load CSV file into DataFrame
df = pd.read_csv('/kaggle/input/nombre-tallacsv/Nombre_Talla.csv')

# Calculate mean, median, and mode of a column
column_name = 'Talla'

mean_value = df[column_name].mean()
median_value = df[column_name].median()
mode_value = df[column_name].mode().values[0]

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
print(df.head)
```
