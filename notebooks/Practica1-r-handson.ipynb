1) Create an array of 10 random integers between 1 and 100, and then calculate the mean and standard deviation of the array.

```r
# Set the seed for reproducibility
set.seed(123)

# Create the array of 10 random integers between 1 and 100
random_array <- sample(1:100, 10)

# Calculate the mean of the array
mean_value <- mean(random_array)

# Calculate the standard deviation of the array
sd_value <- sd(random_array)

# Print the array, mean, and standard deviation
print(random_array)
print(paste("Mean:", mean_value))
print(paste("Standard Deviation:", sd_value))
```

2) Create a 2-dimensional array of 3 rows and 4 columns with random integer values. Then, calculate the sum of each row and column

```r
# Set the seed for reproducibility
set.seed(123)

# Create the 2-dimensional array of 3 rows and 4 columns with random integer values
array <- matrix(sample(1:100, 3*4, replace = TRUE), nrow = 3)

# Calculate the sum of each row
row_sums <- rowSums(array)

# Calculate the sum of each column
col_sums <- colSums(array)

# Print the array, sum of each row, and sum of each column
print(array)
print("Sum of each row:")
print(row_sums)
print("Sum of each column:")
print(col_sums)
```

4) Load a CSV file into a DataFrame and then calculate the mean, median, and mode of one of the columns in the DataFrame.

```r
# Load the CSV file into a DataFrame
df <- read.csv("/kaggle/input/nombre-tallacsv/Nombre_Talla.csv")

# Calculate the mean of a column
mean_value <- mean(df$Talla)

# Calculate the median of a column
median_value <- median(df$Talla)

# Calculate the mode of a column
mode_value <- as.numeric(names(which.max(table(df$Talla))))

# Print the mean, median, and mode
print(paste("Mean:", mean_value))
print(paste("Median:", median_value))
print(paste("Mode:", mode_value))
```

7) Inspect which were the 30 hottest days.

```r
# Sort the data frame by temperature in descending order
sorted_df <- df[order(df$Temp, decreasing = TRUE), ]

# Select the top 30 rows
hottest_days <- head(sorted_df, n = 30)

# Print the 30 hottest days
print(hottest_days)
```

8) Inspect which were the hottest days, but also with the highest ozone values.

10) For each month, obtain the minimum and the maximum temperature registered in Celsius.

```r
# Group the data frame by month and calculate the minimum and maximum temperature
temp_status <- aggregate(TempC ~ Month, data = df, FUN = function(x) c(min = min(x), max = max(x)))

# Print the temperature statistics by month
print(temp_status)
```

11) Obtain the average of the following parameters by month: temperature in celsius, wind, solar radiation and ozone.

```r

# Group the data frame by month and calculate the average for Temp, Wind, Solar.R, and Ozone
averages <- aggregate(cbind(Temp, Wind, Solar.R, Ozone) ~ Month, data = df, FUN = mean)

# Print the averages
print(averages)
```
