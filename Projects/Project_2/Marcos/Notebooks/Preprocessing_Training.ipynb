### Work around null values

```python
# what kind of datatype I'm working with
df.info()
```

We have a categorical variable with null values \
The other ones are numeric values

```python
#identify null values
df.isnull().sum()
```

null values in the categorical variable are 81, it's means we could delete the null values, it is a very small portion of the data set

```python
#delele categoriacal wd registers
df = df.dropna(subset=['wd'])
```

```python
#apply dummies - one-hot encoding
df = pd.get_dummies(df, columns = ['wd'])
df.head()
```

```python
df.isnull().sum()
```

As you can appreciate, the numeric null values are less than 30% of the data, so we can imput, null values

```python
#we imput whit the mean
column_to_imput = ['PM10','SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', ]

for columna in column_to_imput:
    media = df[columna].mean()
    df[columna] = df[columna].fillna(media)
```

```python
#just to confirm, there's no null values
df.isnull().sum()
```

### Working around numerical outliers values



```python
# Select only numeric columns
numeric_variables = df.select_dtypes(include=[np.number]).columns
num_plots = len(numeric_variables)
num_cols = 3
num_rows = math.ceil(num_plots / num_cols)

#draw the plot
plt.figure(figsize=(15, 5 * num_rows))
for i, col_name in enumerate(numeric_variables):
  plt.subplot(num_rows, num_cols, i+1)
  sns.boxplot(y=col_name, data=df)
plt.tight_layout()
plt.show()
```

```python
#cut off 10% of tails
numeric_variables = df.select_dtypes(include=[np.number]).columns

for i in numeric_variables:
  q1 = df[i].quantile(0.1)
  q4 = df[i].quantile(0.9)
  df[i][df[i]<=q1] =q1
  df[i][df[i]>=q4] =q4
```

```python
#draw again to check the tails
# Select only numeric columns
numeric_variables = df.select_dtypes(include=[np.number]).columns

num_plots = len(numeric_variables)
num_cols = 3
num_rows = math.ceil(num_plots / num_cols)

plt.figure(figsize=(15, 5 * num_rows))

for i, col_name in enumerate(numeric_variables):
  plt.subplot(num_rows, num_cols, i+1)
  sns.boxplot(y=col_name, data=df)

plt.tight_layout()
plt.show()
```

Conditions to Air Quality Index

```python
def calc_aqi(row):
  params = ['PM10', 'SO2', 'NO2', 'CO', 'O3']
  max_params_value = row[params].max()
  if 0 <= max_params_value <=50:
    return 'Excellent'
  elif 51 <= max_params_value <=100:
    return 'Good'
  elif 101 <= max_params_value <= 150:
    return 'Slighty Polluted'
  elif 151 <= max_params_value <= 200:
    return 'Lightly Polluted'
  elif 201 <= max_params_value <= 250:
    return 'Moderately Polluted'
  elif 251 <= max_params_value <= 300:
    return 'Heavily Polluted'
  elif max_params_value >=300:
    return 'Severely Polluted'
df
```
